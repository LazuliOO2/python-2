```python
# Depois da execução do código aparece um arquivo csv mostrado os ticket da ações

# Agora iremos obter os dados da criptmoedas.Nesse caso o código é mais robusto e pouco diferente porque para obter até o momento os ticket da ações de criptomoeda é preciso clicka em um botão
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import csv
import time

# URL do site
url = "https://www.infomoney.com.br/cotacoes/cripto/"

# Inicializando o navegador com o WebDriver Manager
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

try:
    # Acessando o site
    driver.get(url)

    # Nome do arquivo CSV
    csv_file = "cripto.csv"

    # Abrindo o arquivo para salvar os dados
    with open(csv_file, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow(["Tickets"])  # Cabeçalho

        while True:
            # Esperar até que os tickets estejam carregados
            try:
                WebDriverWait(driver, 15).until(
                    EC.presence_of_all_elements_located((By.TAG_NAME, "strong"))
                )
            # Exception as e: Isso significa que qualquer tipo de erro (Exception) que acontecer será capturado e armazenado em uma variável chamada e
            except Exception as e:
                print("Os tickets não foram carregados a tempo.")
                break

            # Encontrando os elementos que contêm os tickets
            ticket_elements = driver.find_elements(By.TAG_NAME, "strong")

            # Extraindo o texto dos tickets e salvando no CSV
            # for ticket in ticket_elements: Itera sobre cada elemento na lista ticket_elements (que contém os elementos HTML dos tickets)
            # ticket.text: Extrai o texto de cada elemento
            # if ticket.text.strip(): Verifica se o texto, após remover espaços em branco no início e no fim (strip()), não está vazio. Isso evita adicionar tickets vazios na lista
            tickets = [ticket.text for ticket in ticket_elements if ticket.text.strip()]
            for ticket in tickets:
                writer.writerow([ticket])

            # Esperar até que o botão 'Próximo' esteja disponível e tentar clicar
            try:
                # Verificar se o botão "Próximo" está visível
                next_button = WebDriverWait(driver, 15).until(
                    EC.visibility_of_element_located(
                        (By.CSS_SELECTOR, "button.pagination__btn--next.next")
                    )
                )

                # Mover para o botão antes de clicar (às vezes necessário para garantir interação)
                driver.execute_script("arguments[0].scrollIntoView();", next_button)
                time.sleep(1)  # Pequeno atraso para garantir visibilidade

                # Clicar no botão
                next_button.click()
                time.sleep(3)  # Esperar carregar os próximos dados
            # Exception as e: Isso significa que qualquer tipo de erro (Exception) que acontecer será capturado e armazenado em uma variável chamada e. Essa variável pode ser usada para obter mais informações sobre o erro, como a mensagem de erro
            except Exception as e:
                print("Botão 'Próximo' não encontrado ou não clicável. Paginação concluída.")
                break

    print(f"Tickets salvos no arquivo '{csv_file}'.")
# O bloco finally é usado para garantir que um determinado código seja executado, independentemente de ocorrerem erros ou não
finally:
    # Fechando o navegador
    driver.quit()

```
